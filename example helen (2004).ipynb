```python
import os
import pandas as pd
import numpy as np
import re 
```

```python
#loading in the dataset 
df = pd.read_excel('Table1 2004.xls')
```

```python
#renaming the column names 
df = df.rename(columns={"Table 1": "Bias motivation", "Unnamed: 1": "Incidents", "Unnamed: 2": "Offenses", 
                  "Unnamed: 3": "Victims", "Unnamed: 4": "Known Offenders"})
```

```python
#removing unneeded rows and resetting the index 
df = df.loc[5:30, :].set_index('Bias motivation')
```

```python
#dropping rows that are not needed (the totals)
df = df.drop(df.index[5])
df = df.drop(df.index[12])
df = df.drop(df.index[-4])
df = df.drop(df.index[-6])

df
```

```python
#removing any digits inside the index column 
df = df.rename(index=lambda s: re.sub(r'\d+$', '', s))
```
